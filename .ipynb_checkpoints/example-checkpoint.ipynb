import two functions from our module (simulation.py)

```python
from simulation import genParticles, leapfrog
```

simulation free parameters

```python
L = 2
rand_type = 2 
sig= 0.1
mu = 0.1
N = 10 
storedir = None 
```

generate and visulize particles 

```python
if rand_type > 0: 
    fig, ax = plt.subplots(1, 1, figsize=(5, 5))
    particles = genParticles(2, mu=mu, sig=sig, L=L, N=N)
else: 
    0 
    
ax.scatter(particles[:, 0], particles[:, 1])
ax.set_xlim(-L/2, L/2)
ax.set_ylim(-L/2, L/2)
plt.show()
    
```

run simulation

```python
if storedir is None:
    storedir = 'L{}n{}'.format(L, N)
    
storedir = './data/{}'.format(storedir)
if not os.path.isdir(storedir):
    os.mkdir(storedir)

leapfrog(particles, N=1e3, L=5, theta=0.5, epsilon=1e-3, m_scale=1e3, store=True, path=storedir)
```
